# PCA on MNIST dataset

```python
from sklearn.decomposition import PCA
```

```python
from mnist import MNIST

mndata = MNIST('../data/')

train_x, train_y = mndata.load_training()
test_y, test_y = mndata.load_testing()
```

```python
print(f"#samples: \t{len(train_x)}")
print(f"#features: \t{len(train_x[0])}")
```

```python
pca_num = 784

pca = PCA(n_components = pca_num)
X2D = pca.fit_transform(train_x)

components = (pca.components_.T[:,:pca_num]).T
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

row_max = 5

show_num = 100

f, axarr = plt.subplots( int(show_num/row_max) + 1, row_max)

f.set_figheight(100)
f.set_figwidth(15)

for i in range(0, show_num):
    a = np.reshape( components[i], (28, 28))
    axarr[int(i/row_max),i%row_max].imshow(a)
    
plt.show()
```
